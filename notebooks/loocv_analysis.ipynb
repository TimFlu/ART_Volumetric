# Result analysis

```python
import pandas as pd
path = '../results/loo_selection_results.csv'
loo_selection_results = pd.read_csv(path)
# delete the ' ' around column names
loo_selection_results.columns = [col.strip() for col in loo_selection_results.columns]

# print(loo_selection_results.columns)
loo_selection_results['composite'] = 0.5 * loo_selection_results['recall'] + 0.3 * loo_selection_results['f1'] \
    + 0.1 * loo_selection_results['specificity'] + 0.1 * loo_selection_results['early_detection_rate']
# display(loo_selection_results)

print(f'--------------------------------')
print(f'Best Scores per Variant:')
display(loo_selection_results.groupby('variant').max())
print(f'--------------------------------')
print(f'Top Five Overall Score:')
top_three = loo_selection_results.nlargest(5, 'composite').sort_values(by='avg_early_detection', ascending=False)
display(top_three)

# Select those with the highest f1, then among those select the one with the highest recall, then balanced accuracy, then avg early detection
best_recall = loo_selection_results['recall'].max()
candidates = loo_selection_results[loo_selection_results['recall'] == best_recall]
best_f1 = candidates['f1'].max()
candidates = candidates[candidates['f1'] == best_f1]
best_specificity = candidates['specificity'].max()
candidates = candidates[candidates['specificity'] == best_specificity]
best_early_det = candidates['early_detection_rate'].max()
canditates = candidates[candidates['early_detection_rate'] == best_early_det]
best_early_det = candidates['avg_early_detection'].max()
final_model = candidates[candidates['avg_early_detection'] == best_early_det]
print(f'Final Model Selected:')
display(final_model) 
```
